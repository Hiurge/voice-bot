```python
import pyaudio
import speech_recognition as sr
import os
import random 
import spacy
nlp = spacy.load("en_core_web_sm")

#!sudo apt-get install espeak
#!pip3 install google-search

sr.Microphone.list_microphone_names()
```

```python
def bot_answering(answer):
    os.system('''espeak {}'''.format( answer))
    
def bot_listening():
    r = sr.Recognizer()
    mic = sr.Microphone()
    
    your_message = ''
    with mic as source:
        while not your_message:        
            try:
                r.adjust_for_ambient_noise(source)
                audio = r.listen(source)
                your_message = r.recognize_google(audio)
            except Exception as e:
                print(e)
                bot_answering('Please say it again.')
    return your_message

def bot_logic(your_message):

    msg = your_message.lower()
    print(msg)
    bot_active = True
    
    # some easy entrypoints
    adjs = [w for w in nlp(msg) if w.pos_ == 'ADJ']
    nouns = [w for w in nlp(msg) if w.pos_ == 'NOUN']
    # Argument 'string' has incorrect type (expected str, got spacy.tokens.token.Token)

    # some logic to add here ...
    
    # some test logic:
    
    
    
    # HELLO / KKTHXBYE
    if 'bye bot' in msg:
        bot_active = False
        bot_answer = 'Bye bye {} !'.format(your_name)
    elif 'hello bot' in msg:
        bot_answer = 'Hi {}'.format(your_name)
        
        
        
    # COMMANDS
    elif 'bot tell me about' in msg:
        topic_to_check = msg.split('bot tell me about')[1].split('i look for')[0]
        arguments = msg.split('i look for')[1]
        # logic to find answer ...
        # 
        bot_answer = 'Ha ha ha i cant do that yet, hahaha'
        
    elif 'bot tell me a joke about' in msg:
        joke_about = msg.split('bot tell me a joke about')[1]
        # logic to find a joke ...
        bot_answer = 'lol u mad bro, i am not doing it'
        
    # STICK TO CONVERSATION # DUMB MODE
    elif nouns and adjs:
        topics_list = random.choice([nouns, adjs])
        topic = random.choice(topics_list)
        topic_type = nlp(topic).pos_
        if topic_type == 'ADJ':
            bot_answer = 'What do you find it {}'.format(topic)
        elif topic_type == 'NOUN':
            bot_answer = 'What do you think about {}'.format(topic)
    elif nouns:
        topic = random.choice(nouns)
        bot_answer = 'What do you think about {}'.format(topic)
    elif adjs:
        topic = random.choice(adjs)
        bot_answer = 'What do you find it {}'.format(topic)

    # IDK MODE:
    else:
        bot_answer = 'You said: ' + your_message + ' What did you meant by that?'
    print(bot_answer)
    return bot_answer, bot_active
    
    
    
your_name = 'Mr Fluent'
bot_active = True
log = []
while bot_active == True:
    try:
        
        your_message = bot_listening()
        log.append({'Name': your_name, 'Message': your_message})
        
        bot_answer, bot_active = bot_logic(your_message)
        bot_answering(bot_answer)
        log.append({'Name': 'Bot', 'Message': bot_answer})
        
    except Exception as e:
        print(e)
    
for d in log:
    print('{}: {}'.format(d['Name'], d['Message'])    
```
